## Scheduling


```python
import numpy as np
import dask
import dask.array as da
```

```python
# Local Threads
dask.config.set(scheduler='threads') 
```

```python
dask.config.set
```

```python
dask.config.set(scheduler='processes') 
```


### The .compute method


```python
x = da.arange(100, chunks=10)
x
```

```python
y = (x + 1).sum()
z = (x + 1).mean()
da.compute(y, z)    
```

```python
x.sum().compute()
```

```python
%timeit x.sum().compute(scheduler='threads')
```

```python
%timeit x.sum().compute(scheduler='processes')
```

```python
# set globally
dask.config.set(scheduler='processes')
x.sum().compute()
```

```python
%timeit x.compute(scheduler='processes', num_workers=2)
```

```python
%timeit x.compute(scheduler='threads', num_workers=2)
```

```python
%timeit x.compute(scheduler='threads', n_workers=2, threads_per_worker=2)
```
